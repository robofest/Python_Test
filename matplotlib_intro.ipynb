#plt.plot basics

```python
# %matplotlib inline 
# will make your plot outputs appear and be stored within the notebook.
# not required for Colab
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(x):
    return x**2 - 1
low = -5
high = 5
X = np.arange(low, high, 0.5) # Grid of 0.5 spacing from low to high
Y = func(X) # Evaluate function on xvals
```

```python
print(X)
print(Y)
```

```python
plt.plot(X, Y) # Create line plot with yvals against xvals
plt.show() # Show the figure
```

```python
plt.plot(X, Y, '-v', color="blue") # Create line plot with yvals against xvals
plt.show() # Show the figure
```

```python
rg = np.random.RandomState() # numpy random number generator
print(rg.rand(3))
```

```python
rg = np.random.RandomState() # random number generator using numpy
for marker in ['o', '.', ',', 'x', '+', 'v', '<', 's', 'd']:
    plt.plot(rg.rand(3), rg.rand(3), marker, label="{0}".format(marker))
plt.legend(numpoints=1) # one marker in the legend
plt.xlim(0, 1.5);
```

#### Moving Spines

```python
ax = plt.gca()         # Get Current Axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.plot(X, Y) # Create line plot with yvals against xvals
plt.show() # Show the figure
```

## Scatter Plots with plt.scatter
* More powerful than plt.plot
* Quite similar to plt.plot
* The properties of each individual point (size, face color, edge color, etc.) can be individually controlled or mapped to data

```python
plt.scatter(X, Y, marker='s')
plt.show()
```

```python
plt.scatter(X[0], Y[0], marker='s')
plt.scatter(X[1:-1], Y[1:-1], marker='.')
plt.scatter(X[-1], Y[-1], marker='o')
plt.show()
```

```python
rg = np.random.RandomState(0)
X = rg.randn(500)  # 50 points. What is randn? standard Normal distribution
Y = rg.randn(500)
Colors = rg.rand(500)
#Sizes = 100 * rg.rand(500)

plt.scatter(X, Y, c=Colors, #s=Sizes, 
            alpha=0.8,      # 0.0 transparent through 1.0 opaque
            cmap='viridis') # registered colormap name
plt.colorbar();  # show color scale bar
```

#Plot, Subplot, and Subplot**s** -- 2D contour plots

```python
xlist = np.linspace(low, high, 100) 
ylist = np.linspace(low, high, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = X*X + 2*Y*Y
```

```python
cp = plt.contour(X, Y, Z)
plt.clabel(cp)
plt.title('Contour plot of X^2 + 2Y^2')
plt.xlabel('x0')
plt.ylabel('x1')
plt.show()
```

```python
ax = plt.subplot() # whole figure and an axis (the choice of the name axis is wrong??)
cp = plt.contour(X, Y, Z) # X, Y are from meshgrid
ax.clabel(cp)
ax.set_title("x^2 + 2Y^2")
ax.set_xlabel('x0')
ax.set_ylabel('x1')
plt.show()
```

```python
Z1 = 5*X*X + Y*Y
Z2 = X*X - Y*Y
fig, axes = plt.subplots(2,2)

cp0 = axes[0][0].contour(X, Y, Z)
cp1 = axes[0][1].contour(X, Y, Z1) 
cp2 = axes[1][0].contour(X, Y, Z2)
axes[1][1].plot(X,Y)

axes[0][0].clabel(cp0)
axes[0][0].set_title("X^2 + 2Y^2")
axes[0][0].set_ylabel('x1')
axes[0][0].set_xlabel('x0')

axes[0][1].clabel(cp1)
axes[0][1].set_title("5X^2 + Y^2")
axes[0][1].set_xlabel('x0')
axes[0][1].set_ylabel('x1')

axes[1][0].clabel(cp2)
axes[1][0].set_title("X^2 - Y^2")
axes[1][0].set_xlabel('x0')
axes[1][0].set_ylabel('x1')

fig.tight_layout() # utomatically maintains correct spacing among subplots.
plt.show()
```
